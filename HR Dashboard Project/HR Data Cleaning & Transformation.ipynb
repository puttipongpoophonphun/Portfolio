## Cleaning the Data

```python
HR_Fact = HR_Fact.rename(columns={'Termd':'TermID'})
```

```python
HR_Fact.isnull().sum()
```

```python
# Replace nulls & drop duplicates
HR_Fact.fillna(0,inplace=True)
HR_Fact.drop_duplicates(inplace=True)
```

```python
# Change ManagerID type from float64 to int64
HR_Fact['ManagerID'] = HR_Fact['ManagerID'].astype('int64')
```

## Organize into star schema

```python
# Create Manager Dimension Table
Manager_Dim = pd.DataFrame({'ManagerID': HR_Fact['ManagerID'],
                   'ManagerName': HR_Fact['ManagerName']})
Manager_Dim = Manager_Dim.set_index('ManagerID')
Manager_Dim.drop_duplicates(inplace=True)
Manager_Dim.sort_values('ManagerID',ascending=True,inplace=True)
Manager_Dim
```

```python
# Creating Department Dimension Table
Department_Dim = pd.DataFrame({'DeptID': HR_Fact['DeptID'],
                   'Department': HR_Fact['Department']})
Department_Dim = Department_Dim.set_index('DeptID')
Department_Dim.drop_duplicates(inplace=True)
Department_Dim.sort_values('DeptID',ascending=True,inplace=True)
Department_Dim
```

```python
# Creating Employment Status Dimension Table
EmploymentStatus_Dim = pd.DataFrame({'PositionID': HR_Fact['PositionID'],
                   'Position': HR_Fact['Position']})
EmploymentStatus_Dim = EmploymentStatus_Dim.set_index('PositionID')
EmploymentStatus_Dim.loc[9,'Position'] = 'Data Analyst'
EmploymentStatus_Dim.loc[13,'Position'] = 'IT Manager'
EmploymentStatus_Dim.drop_duplicates(inplace=True)
EmploymentStatus_Dim.sort_values('PositionID',ascending=True,inplace=True)
EmploymentStatus_Dim
```

```python
# Creating Performance Score Dimension Table
PerformanceScore_Dim = pd.DataFrame({'PerfScoreID': HR_Fact['PerfScoreID'],
                   'PerformanceScore': HR_Fact['PerformanceScore']})
PerformanceScore_Dim = PerformanceScore_Dim.set_index('PerfScoreID')
PerformanceScore_Dim.drop_duplicates(inplace=True)
PerformanceScore_Dim.sort_values('PerfScoreID',ascending=True,inplace=True)
PerformanceScore_Dim
```

```python
# Creating Gender Dimension Table
Gender_Dim = pd.DataFrame({'GenderID': HR_Fact['GenderID'],
                   'Sex': HR_Fact['Sex']})
Gender_Dim = Gender_Dim.set_index('GenderID')
Gender_Dim.drop_duplicates(inplace=True)
Gender_Dim.sort_values('GenderID',ascending=True,inplace=True)
Gender_Dim.loc[0,'Sex'] = 'Female'
Gender_Dim.loc[1,'Sex'] = 'Male'
Gender_Dim
```

```python
# Creating Marital Status Dimension Table
MaritalStatus_Dim = pd.DataFrame({'MaritalStatusID': HR_Fact['MaritalStatusID'],
                   'MaritalDesc': HR_Fact['MaritalDesc']})
MaritalStatus_Dim = MaritalStatus_Dim.set_index('MaritalStatusID')
MaritalStatus_Dim.drop_duplicates(inplace=True)
MaritalStatus_Dim.sort_values('MaritalStatusID',ascending=True,inplace=True)
MaritalStatus_Dim
```

```python
# Drop unused columns
HR_Fact = HR_Fact.drop(columns=['ManagerName','Department','EmploymentStatus','Position','Sex','PerformanceScore','MaritalDesc'])
HR_Fact
```

## Export to .csv

```python
HR_Fact.to_csv('output\HR_Fact.csv', index=True)
Department_Dim.to_csv('output\Department_Dim.csv', index=True)
EmploymentStatus_Dim.to_csv('output\EmploymentStatus_Dim.csv', index=True)  
Gender_Dim.to_csv('output\Gender_Dim.csv', index=True)  
Manager_Dim.to_csv('output\Manager_Dim.csv', index=True)  
MaritalStatus_Dim.to_csv('output\MaritalStatus_Dim.csv', index=True)  
```
